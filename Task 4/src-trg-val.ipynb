```python
import pandas as pd
import numpy as np
df = pd.read_excel('data.xlsx')
```

# Edges (src - target - weight)

```python
edge = []

for i in range(len(df['Орг\Тех'])):
    for j in range(1,len(df.columns)):
        if np.isnan(df.iloc[i][j]): 
            continue
        edge.append([df.iloc[i][0],df.columns[j],df.iloc[i][j]])

edges = pd.DataFrame(edge, columns = ['source', 'target', 'weight'])
edges.to_csv('edges.csv', index = False, header = False)
```
